```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
cpath = "Metrics/Manip/Qualitative/manipulation_control_form.csv"
epath = "Metrics/Manip/Qualitative/manipulation_experiment_form.csv"
out_path = "Metrics/Manip/Qualitative"

cdf = pd.read_csv(cpath)
edf = pd.read_csv(epath)

edf
```

```python
def get_dist(s):
    return [
        s[s == 1].shape[0],
        s[s == 2].shape[0],
        s[s == 3].shape[0],
        s[s == 4].shape[0],
        s[s == 5].shape[0]
    ]

get_dist(cdf.iloc[0:, 1])
```

```python
def plot_dist(s, title, filename):
    fig, ax = plt.subplots()
    ax.bar(
        ["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"],
        get_dist(s),
        )
    ax.set_yticks([1, 2, 3, 4, 5, 6])
    ax.set_xlabel("Response")
    ax.set_ylabel("Count")
    ax.tick_params(rotation=90)
    ax.set_title(title)
    plt.savefig(os.path.join(out_path, filename), bbox_inches="tight")
    plt.show()
```

```python
plot_dist(cdf.iloc[0:, 1], "I felt that I was easily able to grab objects.", "q1_control.png")
plot_dist(cdf.iloc[0:, 2], "I found it physically uncomfortable to manipulate\n the objects at times.", "q2_control.png")
plot_dist(cdf.iloc[0:, 3], "I found that I needed to release and re-grab the object\n in order to manipulate it in the way that I wanted.", "q3_control.png")
plot_dist(cdf.iloc[0:, 4], "It was easy to move objects to the location that I wanted.", "q4_control.png")
plot_dist(cdf.iloc[0:, 5], "I found manipulating the objects to be frustrating.", "q5_control.png")
```

```python
plot_dist(edf.iloc[0:, 1], "I felt that I was easily able to grab objects.", "q1_experiment.png")
plot_dist(edf.iloc[0:, 2], "I found it physically uncomfortable to manipulate\n the objects at times.", "q2_experiment.png")
plot_dist(edf.iloc[0:, 3], "I found that I needed to release and re-grab the object\n in order to manipulate it in the way that I wanted.", "q3_experiment.png")
plot_dist(edf.iloc[0:, 4], "It was easy to move objects to the location that I wanted.", "q4_experiment.png")
plot_dist(edf.iloc[0:, 5], "I found manipulating the objects to be frustrating.", "q5_experiment.png")
```

```python
def total_evaluation(s):
    """
    Gets the sum of all the values of the questions for a given response. Positive statements are wieghted positively, negative statements weighted negatively.
    """
    r = s.iloc[1:6]
    acc = 0
    weights = [1, -1, -1, 1, -1]
    for i in range(r.shape[0]):
        acc += int(r.iat[i]) * weights[i]
    return acc

total_evaluation(cdf.iloc[0])
```

```python
ctotals = cdf.apply(total_evaluation, axis=1)
etotals = edf.apply(total_evaluation, axis=1)
print(f"control mean: {ctotals.mean()}\nexperiment mean: {etotals.mean()}")
```
